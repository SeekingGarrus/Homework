```python
df = pd.read_csv("sample_data/california_housing_train.csv")
```

Задание 1. Определить среднюю стоимость дома, где кол-во людей от 0 до 500 (population)

```python
df[df['population'] < 500].median_house_value.mean()
```

Задание 2. Узнать какая максимальная households в зоне минимального значения population

```python
min_population = df.population.min()
df[df['population'] == min_population].households.max()
```

Задание 3. Определить минимальное и максимальное значение median_house_value.

```python
print(df["median_house_value"].max())
print(df["median_house_value"].min())
```

Задание 4. Показать максимальное median_house_value, где median_income = 3.1250

```python
df.loc[(df["median_income"] == 3.1250), ["median_house_value"]].max()
```

Задание 5. Узнать какая максимальная population в зоне минимального значения median_house_value.

```python
min_median_house_value = df.median_house_value.min()
df[df['median_house_value'] == min_median_house_value].population.max()
```
